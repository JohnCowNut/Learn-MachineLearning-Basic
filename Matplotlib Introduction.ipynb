```python
## Matplotlib example workflow
# 0. Import matplotlib and get it ready for ploting in Uipyter
%matplotlib inline
import matplotlib.pyplot as plt
# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) #width , height

# 3. Plot data
ax.plot(x,y)

# 4. Customizze plot
ax.set(title='Simple Plt',
          xlabel="x-axis",
          ylabel ='y-axis')
# 5. Save & show (you save the whole figure)
fig.savefig("./testfig.png")
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}
fig, ax= plt.subplots()
ax.bar(x=list(nut_butter_prices.keys()),height=nut_butter_prices.values()) # x,y
ax.set(title="Thuoq's Nut Butter Store",
        ylabel="Price ($)")
```

```python
# Make some data for histograms and plot it # 
x = np.random.randn(1000)
x
```

```python
fig,ax = plt.subplots()
ax.hist(x) ## hist gram
```

```python
# Two option for subplots
# Subplot option 1 
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                        ncols=2,
                                         figsize=(10,5))

ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```

```python
# Subplots Option2: 
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# Plot to each differenct axis index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(100))
```

### Which one should you ues ("plylot vs matplot lib OO methds"

* When plotting something quickly. okay to use the pyplot
* When plotting some thing more advance , use oO methd

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
#Pyplot method
over_50.plot(kind="scatter",
                x= 'age',
                y ='chol',
            c='target')
```

```python
## OO Method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
                x= 'age',
                y ='chol',
            c='target',
            ax=ax)
ax.set_xlim([45,100])
```

```python
over_50.head()
```

```python
fig,ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
                    y=over_50['chol'],
                    c=over_50['target'])
# Customeize the plot

ax.set(title='Hearte disease and Cholestrol',
       xlabel='Age',
      ylabel='Cholesterol');
## add a legend
ax.legend(*scatter.legend_elements(),title='Target');
## ad a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol, age , thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True)
## add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')
## cusomize ax0
ax0.set(title='Heart Disease and Cholesterol levels',
      
       ylabel ='Cholesterol')
## Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title='Target');
## add a meanline ax0
ax0.axhline(y=over_50["chol"].mean(),
           linestyle='--');
##======================================
## add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])
## cusomize ax1
ax1.set(title='Heart Disease and Thalch levels',
       xlabel='Age',
       ylabel ='Thalach')
## Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title='Target');
## add a meanline ax1
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle='--');
## Add a title to figure
fig.suptitle('Hear Disease Analysis',fontsize=16,fontweight='bold')
```
